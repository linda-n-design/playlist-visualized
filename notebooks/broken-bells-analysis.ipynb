```python
import pandas as pd
import re
import json
```

```python
df = pd.read_excel('../data/Broken Bells Songs.xlsx')
df.head()
```

```python
df = df.drop(columns=['Compilation?', 'Unnamed: 4'], errors='ignore')
df = df.rename(columns={'Version Type': 'Version_Type'})
df.head()
```

```python
def parse_writers(value):
    if pd.isna(value):
        return []
    writers = []
    for entry in str(value).split(';'):
        entry = entry.strip()
        match = re.match(r'^(.+?)\s*\(age\s*(.+?)\)$', entry)
        if match:
            writers.append({
                'name': match.group(1).strip(),
                'age': match.group(2).strip()
            })
        else:
            writers.append({'name': entry, 'age': None})
    return writers

df['Writers'] = df['Writers'].apply(parse_writers)
df[['Song', 'Writers']].head(10)
```

```python
string_cols = ['Song', 'Album', 'Version_Type', 'Notes']
for col in string_cols:
    df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)

df.head()
```

```python
df.to_json(
    '../data/broken-bells-songs.json',
    orient='records',
    force_ascii=False,
    indent=2
)
print("JSON exported successfully!")
```
